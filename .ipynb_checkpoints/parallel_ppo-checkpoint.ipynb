# `ParallelSampler` usage examples

`ParallelSampler` uses [Ray](https://docs.ray.io/en/latest/index.html) as a back-end for multi-thread computing. Next, we will examine different settings of parallel sampling.

`ParallelSampler` takes the following arguments in `__init__()`:

`env`: list -  a list of environments to sample from 

`obs_nlags`: int - the number of lags in addition to current observation

`obs_concat_axis`: int or None - if not None, creates an extra dimension in observation array to concat along, if needed

`obs_expand_axis`: int - axis to concatenate lags along

`obs_padding`: str - ('zeros', 'first') value to pad starting observations with. 

`gpus_share`: float - the total share of GPU memory to allocate for parallel actors (only in case of GPU hosted actors)

# CPU/GPU Hosted Actors - 1 thread vs 5 threads by batch size:

!Note: May not be run if only graphed results are needed

```python
env_name = 'HalfCheetah-v2'
acs_dim = 6
obs_dim = 17
n_actors = 5

sampling_data = {}

for d in ['cuda', 'cpu']:

    # Env for 1 thread sampling:
    env = gym.make(env_name)

    # Envs list for multithread sampling
    envs = [gym.make(env_name) for _ in range(n_actors)]
    for si, env_i in enumerate(envs):
        env_i.seed(np.random.randint(100))

    # Define Samplers:
    sampler = Sampler(env=env)
    parallel_sampler = ParallelSampler(env=envs, gpus_share=0 if d == 'cpu' else 0.9)

    # Actor & Critic
    actor = PPO(
        policy_net=NormalMLP(obs_dim, acs_dim, 
                             nunits=32, nlayers=2,
                             out_activation=torch.nn.Tanh(),
                             init_log_std=-1.0),
        device=torch.device(d),
        learning_rate=2e-4, 
        n_epochs_per_update=20, 
        batch_size=4368, 
        target_kl=1.0, 
        eps=0.2,
        gamma=0.99
    )

    # Running the experiment

    batch_sizes = [1000, 2500, 5000, 10000, 30000]

    sampling_data[d] = {'single_thread': [],
                        f'{n_actors}_threads': []}

    for batch_size in batch_sizes:

        # 1 thread
        print(f'[Device - {d}]: Batch size: {batch_size} - single thread')

        st = datetime.now()
        batch = sampler.sample(actor=actor,
                               n_transitions=batch_size)
        en = datetime.now()

        delta = en-st
        delta = (delta.seconds*int(1e6) + delta.microseconds) / int(1e6)

        sampling_data[d]['single_thread'].append(delta)

        # 10 threads
        print(f'[Device - {d}]: Batch size: {batch_size} - {n_actors} threads')

        st = datetime.now()
        batch = parallel_sampler.sample(actor=actor,
                                        n_transitions=batch_size)
        en = datetime.now()

        delta = en-st
        delta = (delta.seconds*int(1e6) + delta.microseconds) / int(1e6)

        sampling_data[d][f'{n_actors}_threads'].append(delta)

    ray.shutdown()
```

# Plot the Results

```python
for k, v in sampling_data.items():
    
    shift=-0.15
    
    for threads, data in v.items():
        plt.bar(x=np.arange(len(batch_sizes))+shift, 
                height=data, 
                width=0.3, 
                label=threads)
        shift += 2*np.abs(shift)
    
    plt.title(f'{k.upper()} Hosted actor 1 vs {n_actors} threads:')
    plt.legend()
    plt.ylabel('Sample Time (sec.)')
    plt.xticks(np.arange(len(batch_sizes)), 
               [str(b) for b in batch_sizes])
    plt.xlabel('Batch Size')
    plt.show()
```

# Parallelizing Takeaways:

1) __Avoid small batches__. If the tasks are very small, the Ray program can take longer than the equivalent Python program. The issue here is that every task invocation has a non-trivial overhead (e.g., scheduling, inter-process communication, updating the system state) and this overhead dominates the actual time it takes to execute the task.

2) __Sampling from GPU hosted policy is slower__. If possible, the best option is to run sampling phase on CPU to ensure maximum performance, then on a training phase transfer it back to GPU using actor's `.set_device()` method. If update phase is computationally cheap, it may be justified to run purely on a CPU.

# Create training and eval environments

```python
env_name = 'Humanoid-v2'
env = gym.make(env_name)
eval_env = gym.make(env_name)
eval_env.seed(np.random.randint(100))

envs = [gym.make(env_name) for _ in range(6)]
for si, env_i in enumerate(envs):
    env_i.seed(np.random.randint(1000))
```

```python
env.observation_space
```

```python
acs_dim = 17
obs_dim = 376
```

# Define PPO actor:

```python
actor = PPO(
    policy_net=NormalMLP(obs_dim, acs_dim, 
                         nunits=128, nlayers=2,
                         out_activation=torch.nn.Tanh(),
                         acs_scale=0.4,
                         init_log_std=-2.0),
    device=torch.device('cuda'),
    learning_rate=1e-4, 
    n_epochs_per_update=20, 
    batch_size=3276, 
    dataloader_n_workers=0,
    grad_norm_clipping=None,
    target_kl=1.0, 
    eps=0.2,
    ent_coef=0.0,
    gamma=0.995,
    standardize_advantages=True,
    weight_decay=0.0
)
```

# Define GAE-$ \lambda $ critic:

```python
critic = GAE(
    critic_net=VMLP(obs_dim, nlayers=2, nunits=128),
    device=torch.device('cuda'),
    learning_rate=1e-4,
    batch_size=3276,
    gamma=0.995,
    gae_lambda=0.95,
    n_target_updates=1, 
    n_steps_per_update=20
)
```

# Provide actor with critic:

```python
actor.set_critic(critic)
```

# Wrap envs into Sampler:

```python
sampler = ParallelSampler(env=envs, gpus_share=0)
eval_sampler = Sampler(env=eval_env)
```

# Visualize untrained actor

```python
visualize_actor(env=eval_env, actor=actor, nsteps=1000, train_sampling=False)
```

# Run Training and Validation loop:

```python
n_steps = 750
logdir = '/home/nlyubaykin/Desktop/RelaxExamples/ParallelPPO/tensorboard_logs'
run_name = f'parallel_ppo_{env_name}'
logger = SummaryWriter(log_dir=f'{logdir}/{run_name}')
all_logs = []

for step in tqdm(range(n_steps)): 
    
    
    # detach critic for sampling
    actor.set_critic(None)
    
    # transfer actor to CPU for sampling
    actor.set_device(torch.device('cpu'))
    
    # sample training batch
    train_batch = sampler.sample(actor=actor,
                                 n_transitions=30000,
                                 max_path_length=None, 
                                 reset_when_not_done=False,
                                 train_sampling=True)
    
    # transfer actor to GPU for training
    actor.set_device(torch.device('cuda'))
    
    # attach the critic back for training
    actor.set_critic(critic)
    
    # update critic
    critic_logs = critic.update(train_batch)
        
    # update actor
    actor_logs = actor.update(train_batch)
     
    # evaluating the policy
    eval_batch = eval_sampler.sample_n_episodes(actor=actor,
                                                n_episodes=5,
                                                max_path_length=None, 
                                                train_sampling=False)
    # Logging
    train_logs = train_batch.create_logs(prefix='train')
    eval_logs = eval_batch.create_logs(prefix='eval')
    
    logs = {**critic_logs,
            **actor_logs,
            **train_logs,
            **eval_logs}
    all_logs.append(logs)

    for key, value in logs.items():
        logger.add_scalar(key, value, step)
    
logger.flush()

ray.shutdown()
```

```python
%load_ext tensorboard
%tensorboard --logdir /home/nlyubaykin/Desktop/RelaxExamples/ParallelPPO/tensorboard_logs
```

# Plot learning curve

```python
def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w

eval_score = moving_average([l['avg_return/eval'] for l in all_logs], 10)
plt.plot(eval_score)
plt.title(f'Learning curve for {env_name}')
plt.xlabel('Training step')
plt.ylabel('Avg. return')
plt.show()
```
